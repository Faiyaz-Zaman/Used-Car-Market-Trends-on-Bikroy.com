```python
import pandas as pd
```

```python
df = pd.read_csv('secondhand_car_details.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Price"] = df["Price"].str.replace("Tk ", "").str.replace(",", "")

# Convert the Price column to numeric (optional)
df["Price (TK)"] = pd.to_numeric(df["Price"])

df.head()
```

```python
df = df.drop(columns=['Price'])
```

```python
df.head()
```

```python
df["Mileage Drove (KM)"] = df["Mileage Drove"].str.replace(" km", "").str.replace(",", "").astype(int)



# Show the DataFrame with the cleaned columns
df.head()
```

```python
df=df.drop(columns=["Mileage Drove"])
```

```python
df.to_csv("lastsecondhand_car_details.csv", index=False)
```
